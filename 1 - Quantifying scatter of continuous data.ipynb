## Practical examples using NumPy


```python
import numpy as np
from scipy import stats

# The toy dataset put into a NumPy array
data = np.array([64.7, 65.4, 88.3, 64, 71.9])

# Calculate the mean
mean = np.mean(data)
print(f"The mean of the dataset is {mean:.2f}")

# Calculate the sample variance
variance = np.var(data, ddof=1) # ddof=0 by default in NumPy!
print(f"The sample (ddof=1) variance of the dataset is {variance:.3f}")

# Calculate the standard error of the mean
sem = np.std(data, ddof=1) / np.sqrt(len(data))
print(f"The sample (ddof=1) standard error of the mean is {sem:.3f}")

# Calculate the 95% confidence interval (normal distribution)
# The details of the calculation are discussed in a later chapter
ci = stats.sem(data) * stats.norm.ppf((1 + 0.95) / 2., len(data)-1)
lower_bound = np.mean(data) - ci
upper_bound = np.mean(data) + ci
print(f"The corresponding 95% CI is ({lower_bound:.3f}, {upper_bound:.3f})")
```

```python
# Calculate the sample (unbiased) and population standard deviations
std_dev = np.std(data, ddof=1)
pop_std_dev = np.std(data, ddof=0)
print(f"The sample/unbiased (ddof=1) standard deviation is {std_dev:.3f}")
print(f"The population (ddof=0) standard deviation is {pop_std_dev:.3f}")
```

### Geometric mean and standard deviation


```python
# Calculate the geometric mean
gmean = np.exp(np.mean(np.log(data)))
print(f"The geometric mean is {gmean:.3f}")

# Calculate the geometric standard deviation
gstd = np.exp(np.std(np.log(data), ddof=1))
print(f"The geometric standard deviation is {gstd:.4f}")
```

### Weighted statistics


```python
# The dataset
data = np.array([64.7, 65.4, 88.3, 64, 71.9])

# Weights for each data point
weights = np.array([0.1, 0.2, 0.3, 0.1, 0.3])

# Calculate the weighted mean
weighted_mean = np.average(data, weights=weights)
print(f"The weighted mean is {weighted_mean}")
```

## Statistical analysis with other Python packages
### SciPy


```python
# Interquartile range (IQR)
iqr = stats.iqr(data)
print(f"The interquartile range (IQR) is {iqr:.2f}")

# Standard Error of the Mean (SE)
sem = stats.sem(data)
print(f"The standard error of the mean (SE) is {sem:.3f}")

# Descriptive summary
summary = stats.describe(data)
print("Descriptive summary:")
for key, value in summary._asdict().items():
    print(f" {key}: {value}")

# Z-score
z_score = stats.zscore(data)
print(f"The z-scores are {z_score}")

# Coefficient of Variation (CV)
cv = stats.variation(data)
print(f"The coefficient of variation (CV) is {cv:.4f}")

# Geometric mean
gmean = stats.gmean(data)
print(f"The geometric mean is {gmean:.3f}")

# Geometric standard deviation
gstd = stats.gstd(data)
print(f"The geometric standard deviation is {gstd:.4f}")
```

### Pandas


```python
import pandas as pd

df = pd.DataFrame(data, columns=['Data'])

# Calculate the mean
mean = df['Data'].mean()
print(f"The mean of the dataset is {mean:.2f}")

# Calculate the standard deviation
std_dev = df['Data'].std(ddof=1)
print(f"The standard deviation of the dataset is {std_dev:.3f}")

# Calculate the Standard Error of the Mean (SEM)
sem = df['Data'].sem()
print(f"The standard error of the mean (SEM) is {sem:.3f}")

# Calculate the Interquartile Range (IQR)
iqr = df['Data'].quantile(0.75) - df['Data'].quantile(0.25)
print(f"The interquartile range (IQR) is {iqr:.2f}")

# Descriptive summary
summary = df['Data'].describe()
print(f"Descriptive summary:\n{summary.round(3)}")

# Z-score (Series)
z_scores = (df['Data'] - df['Data'].mean()) / df['Data'].std(ddof=1)
print(f"z-scores:\n{z_scores.round(3)}")

# Coefficient of Variation (CV)
cv = df['Data'].std(ddof=1) / df['Data'].mean()
print(f"The coefficient of variation (CV) is {cv:.4f}")
```

### statsmodels


```python
from statsmodels.stats.weightstats import DescrStatsW

# Create a DescrStatsW object
descr_stats = DescrStatsW(data, weights=weights, ddof=0)

# Calculate the weighted mean
mean = descr_stats.mean
print(f"The weighted mean is {mean}")

# Calculate the variance
variance = descr_stats.var
print(f"The variance is {variance:.4f}")

# Calculate the standard deviation
std = descr_stats.std
print(f"The standard deviation is {std:.4f}")
```
