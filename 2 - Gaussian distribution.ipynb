## Visualizing the standard normal distribution
### Gaussian distribution object


```python
import numpy as np
from scipy.stats import norm

# Define the standard normal distribution
μ = 0  # Mean is 0
σ = 1  # Standard deviation is 1

# Define the standard normal distribution used throughout the chapter
gauss = norm(loc=μ, scale=σ)

# Calculate and display the mean and variance
mean, variance = gauss.stats()
print("Distribution:\n", gauss.dist)
print(f"Mean: {mean}, Variance: {variance}")

# Alternative: if we just want to access the values without printing:
# mean = gauss.mean()
# variance = gauss.var()
# std_dev = gauss.std()
```

### The probability density function


```python

import matplotlib.pyplot as plt

# Compute the PDF values for 100 points from -4 to 4 sigmas
x = np.linspace(
    μ - 4 * σ,  # We could also use gauss.ppf(0.0001)
    μ + 4 * σ,  # We could also use gauss.ppf(0.9999)
    num=100)

y = gauss.pdf(x)

# Plot the distribution
fig, axs = plt.subplots(1, 3, figsize=(6, 2.25))

# Fill between 1, 2, and 3 standard deviations
for i, n in enumerate([1, 2, 3]):
    axs[i].plot(x, y, 'k')
    axs[i].fill_between(
        x, y, where=((x > μ-n*σ) & (x < μ+n*σ)), color='orangered')
    axs[i].set_title(f"Area within {n} σ")
    axs[i].set_ylabel('PDF')
    axs[i].set_xlabel('σ')
    axs[i].set_xticks(range(-2, 3, 2))
    axs[i].margins(x=0, y=0)

    axs[i].text(
        -.95, .1,
        # Prints percent data within -n to n standard deviations from the mean
        f"{100 * (1 - 2*(1 - gauss.cdf(n))):.1f}%",
        fontdict={'color': 'white', 'size': 6, 'weight':'bold'})

plt.tight_layout();
```

### The cumulative distribution function


```python

fig, axes = plt.subplots(1, 2, figsize=(6, 3))

# We first plot the CDF function
y = gauss.cdf(x)
axes[0].plot(x, y, color='dodgerblue', lw=3)
axes[0].set_title('Cumulative density function')
axes[0].set_ylabel('CDF')
axes[0].set_xlabel('σ')
axes[0].grid(True)

# And then the PDF with the area filled under the PDF curve up to a given SD
given_SD = 1
y = gauss.pdf(x)
axes[1].plot(x, y, 'k')
axes[1].fill_between(x, y, where=(x <= μ + given_SD*σ), color='dodgerblue',)
axes[1].set_title(fr"Probability up to {given_SD} σ")
axes[1].set_ylabel('Density')
axes[1].set_xlabel('σ')

axes[1].text(
    -1.4, .13,
    f"{100 * gauss.cdf(given_SD):.2f}%",
    fontdict={'color': 'white', 'weight':'bold'},
)
plt.margins(x=0.05, y=0)
plt.tight_layout();
```

```python
print("Cumulative probability in the Gaussian distribution up to:")
for sigma in np.arange(-3, 4):
    print(f"{sigma:>2} σ\tP={gauss.cdf(sigma):.4f}")
```

### The percent point function


```python
# Define the probabilities we want to test
probabilities = [0.5, 0.75, 0.9, 0.95, 0.9772, 0.99, 0.999]

# Calculate the PPFs for each probability in a loop
print("Percent point function in the Gaussian distribution for:")
for p in probabilities:
    ppf = norm.ppf(p)
    print(f"P={p:.4f}: PPF={ppf:>.3f}")
```

### The survival function


```python

fig, axes = plt.subplots(1, 2, figsize=(6, 3))

# We first plot the SF function
y = gauss.sf(x)
axes[0].plot(x, y, color='limegreen', lw=3)
axes[0].set_title('Survival function')
axes[0].set_ylabel('SF')
axes[0].grid(True)
axes[0].set_xlabel('σ')

# And then the PDF with the area filled under the PDF curve from a given SD
given_SD = 1
y = gauss.pdf(x)
axes[1].plot(x, y, 'k')
axes[1].fill_between(x, y, where=(x > μ + given_SD*σ), color='limegreen',)
axes[1].set_title(f"Probability beyond {given_SD} σ")
axes[1].set_ylabel('Density')
axes[1].set_xlabel('σ')

axes[1].annotate(
    xy=(1.5, .05),
    xytext=(2, .2),
    text=f"{100 * gauss.sf(x=given_SD):.2f}%",
    arrowprops={'arrowstyle': '-'},
    weight='bold',
)
plt.margins(x=0.05, y=0)
plt.tight_layout();
```

### The two-tailed SF


```python

fig, axes = plt.subplots(1, 2, figsize=(6, 3))

# We first plot the SF function
y = 2 * (1 - gauss.cdf(np.abs(x)))
axes[0].plot(x, y, color='mediumorchid', lw=3)
axes[0].set_title('Two-tailed survival function')
axes[0].set_ylabel('Two-tailed SF')
axes[0].grid(True)
axes[0].set_xlabel('σ')

# And then the PDF with the area filled under the PDF curve from a given SD
given_SD = 2
y = gauss.pdf(x)
axes[1].plot(x, y, 'k')
axes[1].fill_between(
    x, y,
    where=((x < μ - np.abs(given_SD)*σ) | (x > μ + np.abs(given_SD)*σ)),
    color='mediumorchid',)
axes[1].set_title(f"Probability beyond ±{given_SD} σ")
axes[1].set_ylabel('Density')
axes[1].set_xlabel('σ')


axes[1].text(
    -3.5, .26,
    f"{100 * (2 * (1 - gauss.cdf(given_SD))):.2f}%",
    fontdict={'color': 'black', 'weight':'bold'})
axes[1].arrow(x=-2.8, y=.24, dx=.15, dy=-.21, color='grey', ls=':')
axes[1].arrow(x=-2.4, y=.24, dx=4.1, dy=-.21, color='grey', ls=':')
plt.margins(x=0.05, y=0)
plt.tight_layout();
```
