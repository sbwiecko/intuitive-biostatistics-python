### Using SciPy to obtain critical t-values


```python
from scipy.stats import t

# Define the desired confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate alpha (for two-tailed test)
# Notice that we can use Unicode characters also with Python ;-)
α = 1 - confidence_level

# Define the degrees of freedom (e.g., sample size - 1)
DF_hi = 27

# Calculate the critical t-value (two-tailed)
#t_critical = t.ppf(1 - alpha/2, DF) 
t_critical = t(df=DF_hi).ppf(q=1 - α/2)
# ppf is percent point function (inverse of CDF)

print(f"Two-tailed critical t-value (t*) for a {100*confidence_level}% CI \
with {DF_hi} degrees of freedom = {t_critical:.4f}")
```

```python
# Probability in each tail (alpha/2)
prob_tail = t.cdf(-t_critical, DF_hi)

print(f"Probability in each tail (α/2) for α={100*confidence_level:n}% \
and DF={DF_hi}: {prob_tail:.4f}")
```

```python
import pandas as pd

# Confidence levels
confidence_levels = [0.80, 0.90, 0.95, 0.975, 0.99] # conf_lev = 1 - alpha

# Degrees of freedom
DF_values = list(range(1, 6)) + [10] + list(range(50, 201, 50))

# Create an empty dictionary to store t-values
data = {}

# Calculate t-values and populate dictionary
for df in DF_values:
    t_values=[t.ppf((1+conf_level)/2,df) for conf_level in confidence_levels]
    data[df] = t_values

# Create a DataFrame from the dictionary
df = pd.DataFrame(data, index=confidence_levels)

# Format the table
df.index.name = '1-α'
df.columns.name = 'DF'

# Print the table
print("Two-tailed t* for various DF (columns) and alpha (rows)")
print("=======================================================")
print(df.round(2).to_markdown(numalign='left', stralign='left'))
```

### Sign of the critical t-value


```python
# Parameters
α = 0.05
DF_lo = 11  # Degrees of freedom (n - 1)

# Note that 0.975 - 0.025 = 0.95
t_critical_lower = t.ppf(α/2, DF_lo)      # Lower critical value at q = 0.025
t_critical_upper = t.ppf(1 - α/2, DF_lo)  # Upper critical value at q = 0.975

print(
    f"For {100*(1-α):n}% confidence interval with {DF_lo} degrees of freedom")
print(
    f"Lower t* = {t_critical_lower:.1f} ({100*(1 - confidence_level)/2:.1f}% \
of the area under the curve is to the left of this value)")
print(
    f"Upper t* = {t_critical_upper:.1f} ({100*(1 - confidence_level)/2:.1f}% \
of the area under the curve is to the right of this value)")
```

### Visualizing the t-distribution


```python

import numpy as np
import matplotlib.pyplot as plt

# Calculate critical t-values 
t_critical_lower = t.ppf(α/2, DF_lo)  # Lower critical value at q = 0.025
t_critical_upper = t.ppf(1-α/2, DF_lo)  # Upper critical value at q = 0.975

# Generate x values for the plot
# x = np.linspace(-4, 4, 200)  # Range from -4 to 4 standard deviations
x=np.linspace(
    t(df=DF_lo).ppf(q=.001),  # Percent Point Function
    t(df=DF_lo).ppf(q=.999),
    num=200)

# Calculate t-distribution PDF values
hx = t.pdf(x, DF_lo)

# Create the plot
plt.figure(figsize=(3.5, 3))
plt.plot(x, hx, 'k', lw=2)

# Plot the critical t-values
plt.axvline(
    x=t_critical_lower,
    color='orangered',
    linestyle='--')

plt.axvline(
    x=t_critical_upper,
    color='orangered',
    linestyle='--',
    label=f"t* ({t_critical_upper:.3f})")

# Shade confidence interval areas
plt.fill_between(
    x, hx,
    where=((x <= t_critical_lower) | (x >= t_critical_upper)),
    color='tomato',
    label=f'α ({α})')

# Add labels and title
plt.xlabel('t')
plt.ylabel('Density')
plt.margins(x=0.05, y=0)
plt.yticks([])
plt.title(f"t-distribution with {DF_lo} DFs")
plt.legend(loc=2);
```

### Example of body temperatures


```python
# Population parameters
μ = 36.8
σ = 0.4

# Sample sizes
n1 = 12
n2 = 130
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Create a common normal distribution object
temperature_norm_dist = norm(loc=μ, scale=σ)

# Generate random variate samples with different sizes
sample1 = temperature_norm_dist.rvs(size=n1)
sample2 = temperature_norm_dist.rvs(size=n2)

# Calculate sample means and population standard errors
sample_mean1 = np.mean(sample1)  # or sample1.mean()
sample_mean2 = np.mean(sample2)
pop_sem1 = σ / np.sqrt(n1)
pop_sem2 = σ / np.sqrt(n2)

# Calculate z-critical value for 95% CI (since pop std is known)
z_crit = norm.ppf(0.975)  # ppf gives us z score for 97.5%

# Calculate margin of error
margin_error_z1 = z_crit * pop_sem1
margin_error_z2 = z_crit * pop_sem2

# Calculate confidence intervals
ci_lower_z1 = sample_mean1 - margin_error_z1
ci_upper_z1 = sample_mean1 + margin_error_z1
ci_lower_z2 = sample_mean2 - margin_error_z2
ci_upper_z2 = sample_mean2 + margin_error_z2

# Data for the two subplots
data_statistics_z = (
    (sample_mean1,pop_sem1,margin_error_z1,(ci_lower_z1,ci_upper_z1),n1),
    (sample_mean2,pop_sem2,margin_error_z2,(ci_lower_z2,ci_upper_z2),n2),
)
```

```python

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(6, 3), sharey=True)

# Loop through the data and create plots
for i, (mean, _, margin_error, _, n) in enumerate(data_statistics_z):
    ax = axes[i]  # Get the current subplot axis
    ax.errorbar(
        [4.5],  # Ramdom location for the 95% CI bar
        [mean], yerr=margin_error,
        fmt='r+', capsize=5, label=f'95% CI (n={n})'
    )
    ax.axhline(
        μ, label='μ',
        color='g', linestyle='dashed', linewidth=1)
    ax.scatter(
        range(1, n + 1),
        sample1 if i == 0 else sample2,
        color='b', label='Data points'
    )
    ax.set_xlabel('Data point index')
    ax.set_xlim(0, n+1)
    ax.set_ylabel('Body temperature (°C)')
    ax.set_title(f'Simulation (n = {n})')
    ax.legend()

plt.tight_layout();
```

```python
# Calculate critical t-values (depends on the DF)
t_critical1 = t.ppf(1 - α/2, n1-1)
t_critical2 = t.ppf(1 - α/2, n2-1)

# Calculate **sample** standard deviations
sample_sd1 = np.std(sample1, ddof=1)
sample_sd2 = np.std(sample2, ddof=1)

# Calculate standard error of the mean based on the sample 
#sample_sem1 = np.std(sample1, ddof=1) / np.sqrt(len(sample1))
sample_sem1 = sample_sd1 / np.sqrt(n1) # same as above
sample_sem2 = sample_sd2 / np.sqrt(n2)

# Calculate margin of error
margin_error_t1 = t_critical1 * sample_sem1
margin_error_t2 = t_critical2 * sample_sem2

# Calculate confidence intervals
ci_lower_t1 = sample_mean1 - margin_error_t1
ci_upper_t1 = sample_mean1 + margin_error_t1
ci_lower_t2 = sample_mean2 - margin_error_t2
ci_upper_t2 = sample_mean2 + margin_error_t2

statistics_t = (
    (
        sample_mean1, sample_sem1, margin_error_t1,
        (ci_lower_t1, ci_upper_t1), n1, t_critical1),
    (
        sample_mean2, sample_sem2, margin_error_t2,
        (ci_lower_t2, ci_upper_t2), n2, t_critical2))

# Print the results
print("Calculating confidence intervals using t-distribution\n")

# Loop through the data and print the results
for sample_mean, sample_sem, margin_error, ci, n, t_critical in statistics_t:
    print(f"Sample size n = {n}:")
    print(f"  t* = {t_critical:.4f}")
    print(f"  Sample mean = {sample_mean:.3f}")
    print(f"  Sample SE = {sample_sem:.4f}")
    print(f"  Margin of error: {margin_error:.5f}")
    print(f"  95% confidence interval: ({ci[0]:.4f}, {ci[1]:.4f})\n")
```

```python
# Set random seed for reproducibility
np.random.seed(111)

B = 10000  # Number of replicates
N = len(sample1) # Number of data points in the original sample

# Alternative using standard for loop
# # Initialize an empty list to store the means of the bootstrap replicates
# bs_replicates = []

# # Loop B times to create B bootstrap replicates
# for _ in range(B):
#     # 1. Create a bootstrap sample
#     # Choose N elements randomly from 'sample1' with replacement
#     bootstrap_sample = np.random.choice(sample1, size=N, replace=True)
    
#     # 2. Calculate the mean of the current bootstrap sample
#     mean_of_sample = np.mean(bootstrap_sample)

#     # 3. Append the calculated mean to our list
#     bs_replicates.append(mean_of_sample)
# Transform the list into a NumPy array

# !Make sure to choose exactly N elements randomly **with replacement**
# from the original N-data point set, e.g., size=12 for sample1 set
bs_replicates = np.array([
    np.mean(
        np.random.choice(
            sample1,
            size=N,
            replace=True
        )) for _ in range(B)])

print("Ten first replicates of the resampling:")
print(bs_replicates[:10].round(3))  # print the 10 first replicates
```

### Estimate of the CI


```python
#Calculate the mean, standard error and 95% CI of the bootstrap distribution
bs_m  = np.mean(bs_replicates)
bs_s  = np.std(bs_replicates, ddof=1)
bs_ci = np.percentile(bs_replicates, [2.5, 97.5])

# Print the results
print(f"Bootstrap analysis of sample1 (n={len(sample1)}):")
print(f"  Bootstrap mean = {bs_m:.3f}")
print(f"  Bootstrap standard error = {bs_s:.4f}")
print(f"  Bootstrap 95% percentile: {bs_ci.round(3)}")
```

## Confidence interval of the standard deviation
### Frequentist approach


```python
from scipy.stats import chi2

def std_dev_conf_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for population standard deviation 
    using the frequentist approach.

    Args:
        data (array-like): the sample data
        confidence (float): the desired confidence level (default: 0.95)

    Returns:
        tuple: the lower and upper bounds of the confidence interval.
    """
    α = 1 - confidence
    n = len(data)

    # We first calculate the unbiased (ddof=1) sample standard deviation
    sample_std_dev = np.std(data, ddof=1)

    # The CI for the population SD is derived from
    # the χ2 distribution with DF=n-1
    # Find the critical chi-squared values with PPF
    chi2_lower = chi2.ppf(α / 2, df = n - 1)
    chi2_upper = chi2.ppf(1 - α / 2, df = n - 1)

    # Use sqrt((n - 1) * s**2 / crit_chi2)
    lower_bound = np.sqrt((n - 1) * sample_std_dev**2 / chi2_upper)
    upper_bound = np.sqrt((n - 1) * sample_std_dev**2 / chi2_lower)

    return sample_std_dev, np.array((lower_bound, upper_bound))
```

```python
print("Calculating 95% CI of the standard deviation (χ²):")

for sample in (sample1, sample2):
    n = len(sample)
    s, ci = std_dev_conf_interval(sample)
    print(f" in sample with n={n}, s={s:.3f}, CI:{ci.round(4)}")
```

### Bayesian approach


```python
print("Calculating 95% CI of the standard deviation (`bayes_mvs`):")

for sample in (sample1, sample2):
    n = len(sample)
    # Calculate Bayesian credible interval for SD (α=0.95 for 95% CI)
    result = bayes_mvs(sample, alpha=0.95)
    s, ci = result[2][0], np.array(result[2][1])
    print(f" in sample with n={n}:\ts={s:.3f}, CI={ci.round(4)}")
```
