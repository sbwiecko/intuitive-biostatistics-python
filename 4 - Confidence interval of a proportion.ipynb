## Bernoulli trial and binomial distribution
### Example of a coin flip


```python
import numpy as np

n_single = 1  # Number of coin flips
p_coin = 0.5  # Probability of success = getting heads

# Simulate a single coin flip
flip_result = np.random.binomial(n_single, p_coin)

# Check the result
print("Result of a single random coin flip:")
if flip_result:  # Same as `if flip_result == 1`
    print("Heads")
else:
    print("Tails")
```

### Multiple coin flips


```python
# Parameters of the binomial distribution
n_multiple = 10  # Number of coin flips per experiment

np.random.seed(111)  # Setting seed for reproducibility

# Number of experiments
num_experiments = 10000

# Simulate the experiments
experiments = np.random.binomial(
    n=n_multiple, p=p_coin, size=num_experiments)

# Calculate the mean number of heads
mean_heads = np.mean(experiments)

print(f"Mean number of heads out of 10 flips in \
{num_experiments} experiments: {mean_heads}")
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(3.5, 3))
plt.hist(
    experiments,
    bins=range(n_multiple+2),
    align='left',
    rwidth=0.8,
    color='orangered',
    edgecolor='black',
)
plt.xlabel('Number of heads')
plt.xticks(np.arange(0, 11, 1))
plt.ylabel('Frequency')
plt.title(f"Coin flip simulations (N={num_experiments})");
```

### Probability and frequency


## Computational methods for the binomial distribution
### Binomial distribution


```python
from scipy.stats import binom

binom_dist = binom(n_multiple, p_coin)

print(
    "Frequency of successes in the fifteen first series of coin flips (SciPy):")
print(experiments[:15])

# Calculate the mean, variance and SD of this binomial distribution
print(
    " Mean and variance of the distribution:",
    binom_dist.stats(moments='mv'))
print(" Standard deviation =", round(binom_dist.std(), 4))
```

### Probability mass function
### Lollipop plot


```python

# Calculate the percent point function values
x_start = binom_dist.ppf(q=0.001) # Using the distribution set up previously
x_end = binom_dist.ppf(q=.999)    # Same as x_end=binom_dist.ppf(0.999,n,p)

# Generate x-values
x_lollipop = np.arange(x_start, x_end+1)

# Calculate the probability mass function values
pmf_values = binom_dist.pmf(x_lollipop)

# Plot the PMF using plot and stem methods
_, axes = plt.subplots(
    nrows=1, ncols=2, figsize=(6, 3), sharey=True, sharex=True
)
axes[0].plot(x_lollipop, pmf_values, 'ro', ms=8)

# Lollipop style
# using standard `vlines`
axes[0].vlines(x_lollipop, 0, pmf_values, colors='grey', lw=3, alpha=.5)
# or using the dedicated `stem` function
axes[1].stem(
    x_lollipop, pmf_values, markerfmt="C4s", linefmt="C8--", basefmt=" "
)

for ax in axes:
    ax.set_title('Binomial distribution PMF')
    ax.set_xlabel(f"Number of successes out of {n_multiple}")
    ax.set_ylabel('Probability')
    ax.set_xticks(range(1,10))

plt.tight_layout();
```

### Simulating a clinical trial scenario


```python
# Parameters of the binomial distribution
n_patients = 30       # Number of patients
p_improvement = 0.05  # Probability of a patient showing improvement

# Number of trials
num_trials = 10000

# Simulate the distribution and associated random variates
trials_dist = binom(
    n=n_patients,
    p=p_improvement)

trials_variates = trials_dist.rvs(
    size=num_trials,
    random_state=111  # For reproducibility
)

# Calculate the mean number of improvements
mean_improvements = np.mean(trials_variates)

print(f"With theoretical succes rate of {p_improvement} \
in series of {n_patients} patients:")
print(f" Mean number of improvements in {num_trials} trials: \
{mean_improvements}")
```

### Theoretical vs. empirical estimations


```python
# Calculate the probability of getting exactly 4 improvements
prob_4_improvements = trials_dist.pmf(k=4)

print(f"Theoretical probability of getting exactly \
4 improvements in a single trial = {prob_4_improvements:.3f}")

# Calculate the frequency of getting exactly 4 improvements
freq_4_improvements = (trials_variates == 4).sum()

print(f"Empirical frequency of getting exactly 4 \
improvements in {num_trials} trials = {freq_4_improvements}")
```

```python

# Plot the histogram
plt.figure(figsize=(3.5, 3))
plt.hist(
    trials_variates,
    bins=range(n_patients+2),
    align='left',
    rwidth=0.8,
    color='blue',
    edgecolor='black',
    density=False,)
plt.title(f"Simulations of {num_trials} clinical trials")
plt.xlabel(f"Number of improvements out of {n_patients} patients")
plt.xticks(np.arange(0, 11, 1))
plt.xlim(-.5, 8.5)
plt.ylabel('Cumulative number of trials');
```

### The polling voters example


```python
# Parameters of the binomial distribution
n_voters = 600    # Number of voters in the sample
p_democrat = 0.4  # Probability of a voter being a Democrat

# Simulate the distribution
voters_dist = binom(
    n=n_voters,
    p=p_democrat)

# Number of 'successes' answering the question
k_democrat = int(0.45*n_voters)-1  # Subtract 1 because we want at least 45%

proba_democrats = voters_dist.sf(k_democrat)

print(
    f"With {100*p_democrat:n}% of Democrat voters in samples of {n_voters} \
voters")
print(f"the theoretical probability of getting ≥ {k_democrat} Democrat \
votes = {100*proba_democrats:.4f}%")
```

```python

x = np.arange(  # np.linspace is NOT appropriate for non-continuous data
    voters_dist.ppf(.001),
    voters_dist.ppf(.999)+1, # include the point at the 99.9th percentile
    step=1)
plt.figure(figsize=(3.5, 3))
plt.plot(x, voters_dist.pmf(x), lw=3)
plt.margins(x=0, y=0.01)
plt.title('Binomial distribution PMF')
plt.xlabel('Number of Democrats')
plt.ylabel('Probability');
```

```python
# Number of simulations
num_polls = 10000

# Simulate the samples
voters_variates = voters_dist.rvs(
    size=num_polls,
    random_state=111,  # For reproducibility
)

# Calculate the frequency of getting at least k Democrats
freq_democrats = np.sum(voters_variates >= k_democrat)

# Calculate the empirical probability
empirical_proba_democrats = freq_democrats / num_polls

print(f"Empirical probability of getting at least 45% Democrats \
in the sample = {empirical_proba_democrats}")
```

```python

# plt.hist(samples,bins=range(n+2),align='left',color='blue',density=True)
# density=True normalizes the histogram
plt.figure(figsize=(3.5, 3))
plt.hist(voters_variates, bins=range(n_voters+2), align='left', color='blue')
plt.title(f"Simulation of {num_polls} polls")
plt.xlabel(f"Number of Democrats out of {n_voters} voters")
plt.xlim(195, 285)
plt.ylabel('Counts');
```

## Confidence interval
### Free throw success rate example


```python
# Parameters of the binomial distribution
success_throws = 17  # Numerator: number of successeful throws
num_throws = 20      # Denominator: number of free throw attempts
p_success_throws = success_throws/num_throws  # Success rate

np.random.seed(111) # for reproducibility

for i in range(5):
    print(f"Series of throws #{i+1}: \
{np.random.binomial(num_throws, p_success_throws)} successes")
```

```python

import seaborn as sns  # Let's use another visualization package

num_simulations_throws = 2000

# Simulate the distribution
throws_dist = binom(
    n=num_throws,
    p=p_success_throws)

throws_variates = throws_dist.rvs(
    size=num_simulations_throws,
    random_state=111  # For reproducibility
)

# Plot the histogram
plt.figure(figsize=(3.5, 3))
sns.histplot(
    throws_variates,
    stat='count',
    discrete=True, # Center the bars on the datapoint
    color='yellowgreen',
    edgecolor='black',)
plt.title(
    f"Simulated throws (N={num_simulations_throws})")
plt.xlabel(f"Number of successes out of {num_throws} throws")
plt.ylabel('Count');
```

### Confidence interval of the success rate


### Constructing confidence intervals using classical methods


```python
p_prime = (success_throws + 2) / (num_throws + 4)
W = 2 * (p_prime * (1 - p_prime) / (num_throws + 4))**0.5
ci_inf = p_prime - W
ci_sup = p_prime + W

print(
    f"The 95% confidence interval based on the modified Wald method \nextends \
from {100*ci_inf:.2f}% to {100*ci_sup:.2f}% of successful throws")
```

```python
from scipy.stats import norm
from statsmodels.stats.proportion import proportion_confint as smp_ci

# Calculate the mean and standard deviation of the success rates from the
# empirical variates derived from the previous simulation. Later we'll see
# how to leverage mean and SD directly from the theoretical model.
mean_throws = np.mean(throws_variates)
sd_throws = np.std(throws_variates)

# Calculate the 95% CI using norm.interval
ci_norm = norm.interval(
    confidence=0.95, loc=mean_throws, scale=sd_throws)
print(f"95% CI for successful throws assuming normal distribution (SciPy): \
{np.round(ci_norm)} out of 20")

# Calculate the 95% CI for proportions using proportion_confint
print("95% CI for success rate using statsmodels:")

# alpha=0.05 by default
ci_smp_norm = smp_ci(
    count=success_throws, nobs=num_throws, method='normal')
print(f" 'normal' (default) method: {100*np.round(ci_smp_norm, 4)}%")

ci_smp_binom = smp_ci(
    count=success_throws, nobs=num_throws, method='binom_test')
print(f" 'binom_test' method): {100*np.round(ci_smp_binom, 4)}%")

# An alternative method is Clopper-Pearson (used in GraphPad), called 'beta'
ci_smp_beta = smp_ci(
    count=success_throws, nobs=num_throws, method='beta')
print(f" 'beta' Clopper-Pearson method: {100*np.round(ci_smp_beta, 4)}%")
```

### Prediction interval


```python
pi_low = throws_dist.ppf(0.025)
pi_up  = throws_dist.ppf(0.975)

print(
    f"95% prediction interval in the throw example: [{pi_low:n}, {pi_up:n}]")
```

```python
# Calculate the 95% CI around the median using interval method
pi_binom = throws_dist.interval(confidence=.95)

print(f"95% prediction interval with equal areas around the median: \
{pi_binom} out of 20")
```

### Bootstrapping


```python
# Set random seed for reproducibility
np.random.seed(111)

# Generate the original dataset, encoded as 1 = success, 0 = failure
original_throws = np.random.binomial(1, p_success_throws, num_throws) 

B = 10000  # Number of replicates

# !Make sure to choose exactly N elements randomly **with replacement**
# from the original N-data point set, e.g., size=20 for this example
bs_replicates = np.array([
    np.mean(
        np.random.choice(
            original_throws,
            size=len(original_throws),  # or size=num_throws
            replace=True
        )) for _ in range(B)])
# Could also be achieved using a standard Python loop

print("Mean success rates from the first ten bootstrap replicates:")
print(bs_replicates[:10])
```

```python
# Calculate the 95% confidence interval using np.percentile
bs_ci = np.percentile(bs_replicates, [2.5, 97.5])

# Print the results
print(f"Bootstrap 95% CI estimate = {bs_ci.round(2)}")
```

```python

# Plot the histogram of the bootstrap distribution of correlation coefficients
plt.figure(figsize=(3.5, 3))
plt.hist(
    bs_replicates,
    density=False,
    bins=10,
    color='gold',
    label='Bootstrap proportions (p)')

# Annotate the observed mean difference
plt.axvline(
    x=p_success_throws,
    color='cornflowerblue',
    linestyle='-', lw=2,
    label=f'Population proportion ({p_success_throws:.2f})')

# Add lines for the confidence interval
plt.axvline(
    x=bs_ci[0],
    color='red',
    linestyle='--',
    label=f'2.5th ({bs_ci[0]:.2f})')
plt.axvline(
    x=bs_ci[1],
    color='red',
    linestyle='-.',
    label=f'97.5th ({bs_ci[1]:.2f})')

# Add labels and title
plt.xlabel('Proportion (p)')
plt.ylabel('Frequency')
plt.title(f"Bootstrap proportions")
plt.legend();
```

### What does 95% confidence really mean?


```python
# Parameters of the binomial distribution
# Denominator: number of balls drew in each trial
num_balls = 15
success_balls = .25*num_balls  # Numerator: number of red balls per trial
p_success_balls = success_balls/num_balls  # Success rate

# Number of simulations
num_simulations_balls = 100

# Simulate the probability rates with a theoretical binomial distribution
balls_dist = binom(n=num_balls, p=p_success_balls)
balls_variates = balls_dist.rvs(size=num_simulations_balls, random_state=111)

print(r"95% confidence interval in the simulation of red balls:")
# Calculate the 95% CI empirically
ci_balls_empir = np.percentile(balls_variates, [2.5, 97.5])
print(f" empirical: {ci_balls_empir.round(2)} out of 15")

# Calculate the 95% CI assuming nromal distribution
ci_balls_norm = norm.interval(
    confidence=0.95, loc=balls_dist.mean(), scale=balls_dist.std())
print(f" assuming normal distribution: \
{np.round(ci_balls_norm, 2)} out of 15")
```

```python
# Discrete number of successes and trials
red_balls = 4     # Number of red balls drawn
total_balls = 15  # Total number of balls drawn

# Calculate the 95% confidence interval for the proportion
ci_smpci_balls = smp_ci(
    count=red_balls,
    nobs=total_balls,
    alpha=0.05,  # default
    method='binom_test')

print(
    f"95% CI for the proportion of red balls out of 15 (using 'binom_test'): \
{100*np.round(ci_smpci_balls, 4)}%")
```

```python

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(6.5,3))

# Loop over the data and create error bars. Remind that the random 
# variates is the number of successes in each individual trial.
for i, count in enumerate(balls_variates):
    # Calculate the confidence interval
    ci = smp_ci(count, num_balls, method='binom_test')

    # Determine the color based on whether
    # the interval contains the true proportion
    color = (
        'red' if ci[0] > p_success_balls or ci[1] < p_success_balls else 'grey')

    # Plot the error bar
    ax.errorbar(
        i, count / num_balls,
        yerr=[[count / num_balls - ci[0]], [ci[1] - count / num_balls]],
        color=color, fmt='.')

plt.axhline(p_success_balls, ls='--', lw=3, color='cornflowerblue')
plt.title('95% CIs for proportion of red balls')
plt.xlabel('Trial number')
plt.xticks([])
plt.xlim(.5, 100 + .5)
plt.ylabel('Proportion of red balls')
sns.despine();
```
