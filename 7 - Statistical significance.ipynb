### Variability of P values under a true effect


```python
import numpy as np
from scipy.stats import ttest_ind

# For reproducibility
np.random.seed(111)

# Parameters
sample_size         = 10
population_std_dev  = 5.0
mean_difference     = 5.0
num_simulations     = 1000

# Store P values in a list
p_values = []

# Run simulations
for _ in range(num_simulations):
    # Sample from two populations with different means
    data1 = np.random.normal(0, population_std_dev, sample_size)
    data2 = np.random.normal(
        mean_difference, population_std_dev, sample_size)
    
    # Perform t-test
    _, p_value = ttest_ind(data1, data2)
    p_values.append(p_value)

# Print parameters
print("Simulation parameters:")
print(f" Sample size n={sample_size}")
print(f" Population standard deviation σ={population_std_dev}")
print(f" Mean difference (known)={mean_difference}")
print(f" Number of simulations: {num_simulations}")
print("-" * 36) # Add a seperator for clarity

# Analyze P values
print(
    "Proportion of P values < 0.05:",
    np.mean(np.array(p_values) < 0.05))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(111)

α=0.05  # Threshold significance

# Simulation parameters
num_simulations = 2500

# Store P values
p_values = []

# Run simulations
for _ in range(num_simulations):
    data1 = np.random.normal(0, population_std_dev, sample_size)
    data2 = np.random.normal(
        mean_difference, population_std_dev, sample_size)
    _, p_value = ttest_ind(data1, data2)
    p_values.append(p_value)

# Calculate 95% percentiles of the distribution of P values
percentiles = np.percentile(p_values, [2.5, 97.5])

# Plot the distribution of P values
sns.set_theme(style="whitegrid")
plt.figure(figsize=(3.5, 3))

sns.histplot(
    p_values,
    kde=False,
    log_scale=True,  # Histogram with log scale
)
# Vertical line at P=0.05 (the actual α)
plt.axvline(α, color='r', linestyle='dashed', linewidth=1.5, label=f"{α=}")

# Add percentile lines and annotations
for p in percentiles:
    plt.axvline(p, c='k', linestyle='dotted', linewidth=1)
    plt.text(p, 10, f"{p:.5f}", rotation=90, va='bottom', ha='center')

plt.xlabel("P value (log scale)")
plt.ylabel("Frequency (log scale)")
plt.title(f"P value distribution ({num_simulations} simulations)")
plt.legend()
sns.despine();
```

### The risk of false positives under the null hypothesis


```python

# Simulation parameters
n = 10   # Sample size per group
σ = 5.0  # Common standard deviation
μ = 10   # Common mean for both populations

num_simulations = 2500

# For reproducibility
np.random.seed(111)

# Store P values in a list
p_values = []

# Run simulations
for _ in range(num_simulations):
    pop_1 = np.random.normal(loc=μ, scale=σ, size=n)
    pop_2 = np.random.normal(loc=μ, scale=σ, size=n)
    _, p_value = ttest_ind(pop_1, pop_2)
    p_values.append(p_value)

# Calculate significance proportion
signif_proportion = np.mean(np.array(p_values) <= α)

# Plotting
# Seaborn theming remains
plt.figure(figsize=(3.5, 3))
sns.histplot(p_values, kde=False, log_scale=True)
plt.axvline(
    α,
    color='r', linestyle='dashed', linewidth=1.5,
    label=f"{α=}")
plt.xlabel("P value (log scale)")
plt.ylabel("Frequency (log scale)")
plt.title(f"P value distribution ({num_simulations} simulations)")
plt.legend()
sns.despine();
```

### Influence of sample size


```python

from scipy.stats import t

# Simulation parameters
μ1, μ2 = 10, 15  # Group means (fixed difference)
mean_diff = μ2 - μ1
σ = 5.0

p_values = []

for n in range(3, 31):   # Varying sample sizes (n)
    # Calculate t-statistic (see future chapter on un paired t-tests)
    t_stat = mean_diff / (σ * (2 / n)**0.5)
    # Two-tailed P value
    p_value = 2 * (1 - t.cdf(x=abs(t_stat), df = 2*n - 2))
    p_values.append(p_value)

# Plotting
sns.set_theme(style="white")
plt.figure(figsize=(3.5, 3))
sns.lineplot(x=list(range(3, 31)), y=p_values, lw=2, color="blue")
plt.xlabel("Sample size (per group)")
plt.ylabel("P value")
plt.title("P value distribution")
plt.axhline(
    α,
    color='r', linestyle='dashed', linewidth=1.5,
    label=f"{α=}")
plt.legend();
```

### The danger of ad hoc sample size decisions


```python

# Simulation parameters
sample_size_max = 75
num_simulations = 5
common_mean = 10
common_std_dev = 1.0

# Seaborn color palette (biostatistics can also be beautiful)
palette = sns.color_palette("viridis", n_colors=num_simulations)

np.random.seed(111)

# Plotting setup
plt.figure(figsize=(3.5, 3))

# Run simulations and plot
for sim_num in range(num_simulations):
    # Generate data for two populations with the same mean and SD
    pop_1 = np.random.normal(
        loc=common_mean, scale=common_std_dev, size=sample_size_max)
    pop_2 = np.random.normal(
        loc=common_mean, scale=common_std_dev, size=sample_size_max)

    p_values = []
    sample_sizes = list(range(5, sample_size_max + 1))

    # Iterate over increasing sample sizes
    for n in sample_sizes:
        _, p_value = ttest_ind(pop_1[:n], pop_2[:n])
        p_values.append(p_value)

    # Plot for this simulation using the Seaborn palette
    sns.lineplot(
        x=sample_sizes, y=p_values,
        color=palette[sim_num])

# Add labels and title
plt.axhline(
    0.05,
    color='r', linestyle='dashed', linewidth=1.5,
    label=f"{α=}")
plt.xlabel("Sample size (per group)")
plt.ylabel("P value (log scale)")
plt.title("Cumulative P values under H0")
plt.yscale('log')
plt.ylim((0.01, 1))
plt.xlim((0, sample_size_max + 5))
plt.legend()
sns.despine();
```

```python

from sklearn.metrics import roc_curve, auc

# Generate an imaginary dataset
np.random.seed(42)
y_true = np.random.randint(0, 2, 1000)  # True labels (0 or 1)
y_scores = np.random.rand(1000)  # Predicted probabilities (scores)

# Generate scores based on a normal distribution
# with different means for the 2 classes
y_scores = np.zeros(1000)
for i in range(1000):
    if y_true[i] == 1:
        y_scores[i] = np.random.normal(0.7,0.2)
        # class one scores are higher on average
    else:
        y_scores[i] = np.random.normal(0.3, 0.2)
        # class zero scores are lower on average

# Calculate the ROC curve
plt.figure(figsize=(3.5, 3))

fpr, tpr, thresholds_real = roc_curve(
    y_true, y_scores, drop_intermediate=True)

# Calculate the AUC
roc_auc_real = auc(fpr, tpr)

plt.plot(
    fpr, tpr,
    color='red', lw=2,
    label='ROC curve (AUC = %0.2f)' % roc_auc_real)
plt.plot(
    [0, 1], [0, 1],
    color='gray', lw=2, linestyle='--',
    label='Random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False positive rate (FPR)')
plt.ylabel('True positive rate (TPR)')
plt.title('Example of a ROC curve')
plt.legend(loc='lower right')
plt.grid(True);
```
